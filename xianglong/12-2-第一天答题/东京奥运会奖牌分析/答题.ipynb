```python

from pyecharts.charts import Timeline, Bar, Pie
```

```python
data = {}
for i in df5.index:
    data[i] = []
    for ii in df5.columns:
        data[i].append({
            'name': ii,
            'value': int(df5[ii][i])
        })
list(df5.columns)
```

```python
def myBar(day: str) -> Bar:
    bar = (
        Bar()
        .add_xaxis(xaxis_data=list(df5.columns))
        .add_yaxis(
        series_name='奖牌总数',
            y_axis=data[day],
            is_selected=True,
            label_opts=opts.LabelOpts(is_show=False)
        )
    )
    pie = (
        Pie()
        .add(
            series_name='奖牌占比',
            data_pair=[
                [i['name'], i['value']] for i in data[day]
            ],
            center=['75%', '35%'],
            radius='28%'
        )
        .set_series_opts(tooltip_opts=opts.TooltipOpts(is_show=True, trigger='item'))
    )
    return bar.overlap(pie)
```

```python
timeline = Timeline(init_opts=opts.InitOpts(width='1600px', height='800px'))

for i in df5.index:
    timeline.add(myBar(i), time_point=str(i))

timeline.add_schema(is_auto_play=True, play_interval=1000)
timeline.render('finance.html')
```
