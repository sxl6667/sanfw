### 21 从剪贴板读取数据

<br>

打开当前目录下 `Titanic.txt` 文件，全选并复制。

现在直接从剪贴板读取数据。

```python
import win32con
import win32clipboard as w

def get_text():
    w.OpenClipboard()
    d = w.GetClipboardData(win32con.CF_TEXT)
    w.CloseClipboard()
    return d.decode('utf-8')

def set_text(aString):
    w.OpenClipboard()
    w.EmptyClipboard()
    w.SetClipboardData(win32con.CF_TEXT, aString)
    w.CloseClipboard()
    
ss = get_text()
ss
set_text('ss = get_text()')
```

```python
class text():
    s = None
    def __init__(self):
        pass
    def __enter__(self):
        w.OpenClipboard()
        self.s = w.GetClipboardData(win32con.CF_TEXT)
        return self.s.decode('utf-8')
    def __exit__(self, exc_type, exc_val, exc_tb):
        w.CloseClipboard()
```

```python
with text() as s:
    print(s)
```

```python
with text() as s:
    data = [i.split('\t') for i in s.split('\n')]
    my_dict = {}
    for i in range(len(data[0])):
        my_dict[data[0][i]] = []
        for ii in range(len(data)):
            my_dict[data[0][i]].append(data[ii][i])
        my_dict[data[0][i]] = my_dict[data[0][i]][1:]
    df = DataFrame(my_dict)
    print(df)
```
